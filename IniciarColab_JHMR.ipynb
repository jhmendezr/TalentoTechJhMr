```python
#ventas por region
ventas_región=df_ventas.groupby('Region')['Ventas'].sum()
print(ventas_región)
```

```python
#Ventas realizadas por David
Ventas_david=df_ventas[df_ventas['Vendedor']=='David']
print(Ventas_david)
```

```python
# ventas mayores a 10000 por region, si cambias el ascending dependiendo el true
# or false puedes ponerlo de menor a mayor y viceversa
df_filtrada=df_ventas[df_ventas['Ventas']>10000].sort_values(by='Ventas',ascending=False)
print(df_filtrada)
```

```python
#Eliminar columna vendedor
df_eliminado=df_ventas.drop(columns='Vendedor')
print(df_eliminado)
```

```python
#ventas por region
ventas_región=df_ventas.groupby('Region')['Ventas'].sum()
print(ventas_región)
```

```python
#Calcular el promedio de ventas por región con dos decimales
ventas_por_producto=df_ventas.groupby('Producto')['Ventas'].mean().round(2)
print(ventas_por_producto)
```

```python
#Ordenar el dataframe por ventas en orden ascendente
df_ordenado=df_ventas.sort_values(by='Ventas')
print(df_ordenado)
```
